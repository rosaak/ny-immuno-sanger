### how many alignments 

```python
print("query_name|pattern|vh_header|vh_seq|number_of_alignments")
for k,v in df.to_dict().get('CDR-H3').items():
    for vh_k, vh_v in vh_data_ss.items():
        for rec in SeqIO.parse(str(vh_v), "abi"):
            print(f"{k}|{v}|{vh_k}|{rec.seq.upper()[0:10]}...|", end='')
            aligner = Align.PairwiseAligner()
            aligner.gap_score = -1
            aligner.extend_gap_score = -1
            aligner.mismatch_score = -1
            aligner.mode = "local"
            alignments = aligner.align(rec.seq.upper()[0:50], v)
            print(f"{len(alignments)}")
            
            # iterating through all the allignments
            # for e,aln in enumerate(alignments, start=1):
            #     print(e, aln.query, aln.target[0:10], aln.shape, aln.score)
            
            # checking the first alignment
            alignment = alignments[0]
            print(f"-> {alignment}", end="")
            print(f"{alignment.query}|{alignment.target[0:50]}|{alignment.score}\n")
    print()
```

### note

- no perfect alignment
